```python
import requests
import json
import pandas as pd

# Set the API endpoint and API key
data_url = 'https://api.bls.gov/publicAPI/v2/timeseries/data/'
api_key = '1315085c23bf46318cdba4ea8abfbfaf'

# series IDs
series_ids = ['MLUMS00NN0001003', 'LNS14000000', 'LNS13000000']

# Set the request parameters for the API
start_year = '2010'
end_year = '2023'
data = json.dumps({"startyear": start_year, "endyear": end_year, "registrationkey": api_key})
headers = {'Content-type': 'application/json'}

# Loop over each series ID, retrieve the data, and save it to a separate CSV file
for series_id in series_ids:
    # request parameters for the API
    data = json.dumps({"seriesid": [series_id], "startyear": start_year, "endyear": end_year,
                       "registrationkey": api_key})
    
    # Send the POST request to the API endpoint and retrieve the data
    response = requests.post(data_url, data=data, headers=headers)
    json_data = json.loads(response.text)
    results = json_data['Results']['series'][0]['data']

    # Convert Pandas DataFrame and set the index to the date
    df = pd.DataFrame(results)
    df['value'] = pd.to_numeric(df['value'])
    df['date'] = pd.to_datetime(df['year'] + '-' + df['period'])
    df = df.set_index('date')
    df = df[['value']]

    # Save the DataFrame csv
    file_name = series_id + '.csv'
    df.to_csv(file_name)

```

```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt

# URL and API key
url = 'https://api.bls.gov/publicAPI/v2/timeseries/data/'
api_key = '1315085c23bf46318cdba4ea8abfbfaf'

# Start and end year for the data
start_year = '2010'
end_year = '2023'

# Set the industry codes for the data we want to retrieve
industry_codes = ['CES0000000001', 'CES2000000001', 'CES4200000001', 'CES5000000001']

# Set the request parameters
data = json.dumps({
    "seriesid": industry_codes,
    "startyear": start_year,
    "endyear": end_year,
    "registrationkey": api_key
})

# Request headers
headers = {'Content-type': 'application/json'}

# Retrieve the data
response = requests.post(url, data=data, headers=headers)
json_data = json.loads(response.text)

# Extract relevant data from the JSON response
results = json_data['Results']['series']
data = []
for result in results:
    series_id = result['seriesID']
    series_data = result['data']
    for point in series_data:
        year = point['year']
        period = point['period']
        value = point['value']
        data.append((series_id, year, period, value))

# Convert data to a DataFrame
df = pd.DataFrame(data, columns=['Series ID', 'Year', 'Period', 'Value'])
df['Date'] = pd.to_datetime(df['Year'] + '-' + df['Period'].str.extract('(\d+)', expand=False) + '-01')

# Pivot the DataFrame to have the industry codes as columns
df_pivot = df.pivot(columns='Series ID', values='Value')

# Plot the data
fig, ax = plt.subplots(figsize=(10, 6))

df_pivot.plot(ax=ax)

ax.set_xlabel('Date')
ax.set_ylabel('Number of Employees')
ax.set_title('Employment in Different Industries (Seasonally Adjusted)')

plt.show()

```
