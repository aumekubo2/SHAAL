```python
import requests
import json
import pandas as pd

# Set the API endpoint and API key
data_url = 'https://api.bls.gov/publicAPI/v2/timeseries/data/'
api_key = '1315085c23bf46318cdba4ea8abfbfaf'

# series IDs
series_ids = ['MLUMS00NN0001003', 'LNS14000000', 'LNS13000000']

# Set the request parameters for the API
start_year = '2010'
end_year = '2023'
data = json.dumps({"startyear": start_year, "endyear": end_year, "registrationkey": api_key})
headers = {'Content-type': 'application/json'}

# Loop over each series ID, retrieve the data, and save it to a separate CSV file
for series_id in series_ids:
    # request parameters for the API
    data = json.dumps({"seriesid": [series_id], "startyear": start_year, "endyear": end_year,
                       "registrationkey": api_key})
    
    # Send the POST request to the API endpoint and retrieve the data
    response = requests.post(data_url, data=data, headers=headers)
    json_data = json.loads(response.text)
    results = json_data['Results']['series'][0]['data']

    # Convert Pandas DataFrame and set the index to the date
    df = pd.DataFrame(results)
    df['value'] = pd.to_numeric(df['value'])
    df['date'] = pd.to_datetime(df['year'] + '-' + df['period'])
    df = df.set_index('date')
    df = df[['value']]

    # Save the DataFrame csv
    file_name = series_id + '.csv'
    df.to_csv(file_name)

```

```python
import requests
import json
import pandas as pd

# Set the API endpoint and API key
data_url = 'https://api.bls.gov/publicAPI/v2/timeseries/data/'
api_key = '1315085c23bf46318cdba4ea8abfbfaf'

# Set request headers and data
headers = {'Content-type': 'application/json'}

#create empty DataFrame
series_ids = ['BDS0000000000000000120004LQ5','BDS0000000000000000110001LQ5','BDS0000000000000000110004LQ5','BDS0000000000000000120001LQ5','SMU19197802023800001', 'NCU5306633300003', 'LAUCN281070000000003', 'JLU00000000HIL', 'LNS14000000', 'LNS13000000', 'CUUR0000SA0']
df = pd.DataFrame(columns=['series_id', 'year', 'period', 'value'])

# Loop through series IDs and make API request
for series_id in series_ids:
    data = json.dumps({
        "seriesid": [series_id],
        "startyear": "2010",
        "endyear": "2023",
        "registrationkey": api_key
    })
    response = requests.post(data_url, headers=headers, data=data)
    json_data = json.loads(response.text)
    
    # Extract data for series and append to DataFrame
    series_data = json_data['Results']['series'][0]['data']
    df_temp = pd.DataFrame(series_data, columns=['year', 'period', 'value'])
    df_temp['series_id'] = series_id
    df = pd.concat([df, df_temp])

# Reset index and print
df.reset_index(drop=True, inplace=True)
print(df.head())
print(df.tail(10))
```

```python

%matplotlib inline

base_url = 'https://api.bls.gov/publicAPI/v2/timeseries/data/'
api_key = '1315085c23bf46318cdba4ea8abfbfaf'
series_dict = {
    'LNS11300012': '16-19 years',
    'LNS11300036': '20-24 years'
}
start_year = '2010'
end_year = '2023'

headers = {'Content-type': 'application/json'}

data = json.dumps({
    "seriesid": list(series_dict.keys()),
    "startyear": start_year,
    "endyear": end_year,
    "registrationkey": api_key
})

response = requests.post(base_url, headers=headers, data=data)
json_data = response.json()['Results']['series']

date_list = [f"{i['year']}-{i['period'][1:]}-01" for i in json_data[0]['data']]

df = pd.DataFrame()

for s in json_data:
    df[series_dict[s['seriesID']]] = pd.Series(
        index = pd.to_datetime(date_list),
        data = [i['value'] for i in s['data']]
        ).astype(float).iloc[::-1]

df['mean'] = df.mean(axis=1)

plt.figure(figsize=(10, 6))
plt.plot(df.index, df['16-19 years'], linewidth=2, linestyle='-', label='16-19 years')
plt.plot(df.index, df['20-24 years'], linewidth=2, linestyle='-', label='20-24 years')
plt.plot(df.index, df['mean'], linewidth=2, linestyle='--', label='Mean')
plt.ylim([0, 100])
plt.legend()
plt.title('Labor Force Participation Rate by Age Group')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.show()

```

```python

# Set the API endpoint and API key
data_url = 'https://api.bls.gov/publicAPI/v2/timeseries/data/'
api_key = '1315085c23bf46318cdba4ea8abfbfaf'

# Set request headers and data for CPI
headers_cpi = {'Content-type': 'application/json'}
data_cpi = json.dumps({
    "seriesid": ["CUUR0000SA0"],
    "startyear": "2010",
    "endyear": "2023",
    "registrationkey": api_key
})
response_cpi = requests.post(data_url, headers=headers_cpi, data=data_cpi)
json_data_cpi = json.loads(response_cpi.text)

# Set request headers and data for NAICS 10
headers_naics10 = {'Content-type': 'application/json'}
data_naics10 = json.dumps({
    "seriesid": ["ENU0423200510"], #seriesid for NAICS10 sector employment
    "startyear": "2010",
    "endyear": "2023",
    "registrationkey": api_key
})
response_naics10 = requests.post(data_url, headers=headers_naics10, data=data_naics10)
json_data_naics10 = json.loads(response_naics10.text)

# Extract data for CPI and NAICS 10
cpi_data = json_data_cpi['Results']['series'][0]['data']
naics10_data = json_data_naics10['Results']['series'][0]['data']

# Create DataFrames for CPI and NAICS 10
df_cpi = pd.DataFrame(cpi_data, columns=['year', 'period', 'value_cpi'])
df_naics10 = pd.DataFrame(naics10_data, columns=['year', 'period', 'value_naics10'])

# Merge DataFrames
df_merged = pd.merge(df_cpi, df_naics10, on=['year', 'period'])

# Convert data types
df_merged['value_cpi'] = pd.to_numeric(df_merged['value_cpi'])
df_merged['value_naics10'] = pd.to_numeric(df_merged['value_naics10'])

# Create scatter plot of CPI and NAICS 10
plt.scatter(df_merged['value_cpi'], df_merged['value_naics10'])
plt.xlabel('CPI')
plt.ylabel('NAICS 10 Employment')
plt.title('Correlation between CPI and NAICS 10 Employment')
plt.figure(figsize=(8,6))

plt.show()
plt.show()

```
